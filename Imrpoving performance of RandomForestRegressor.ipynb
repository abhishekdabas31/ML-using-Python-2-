# Dealing with missing values in Data


1) Drop columwith missing values
Unless most values in the dropped columns are missing, the model loses access to a lot of (potentially useful!) i
2)Imputation
Fillimg in the missing values with some number. In most cases, we fill in the mean value along each column.
3)An Extension To Imputation
Impute the missing values, as the 2nd step but, additionally, for each column with missing entries in the original dataset, we add a new column that shows the location of the imputed entries (with TRUE OR FALSE)

Testing all 3 methods:-


```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
# reading data:
X_full = pd.read_csv('C:\\Users\\Abhishek\\Desktop\\My_Projects\\ML-using-Python-2-\\train.csv', index_col='Id')
X_test_full = pd.read_csv('C:\\Users\\Abhishek\\Desktop\\My_Projects\\ML-using-Python-2-\\test.csv', index_col='Id')
```

```python
X_full.dropna(axis=0, subset = ['SalePrice'] , inplace= True)
y=X_full.SalePrice
X_full.drop(['SalePrice'], axis=1, inplace=True)
```

# For keeping things simple well just use the numerical columns

```python
X= X_full.select_dtypes(exclude=['object'])
X_test= X_test_full.select_dtypes(exclude=['object'])
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(X,y, random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

# Function for comparing different approaches
def score_dataset(X_train, X_valid, y_train, y_valid):
    model = RandomForestRegressor(n_estimators=100, random_state=0)
    model.fit(X_train, y_train)
    preds = model.predict(X_valid)
    return mean_absolute_error(y_valid, preds)
```

# 1st approach of dropping values

Find the columns that need to be dropped

```python
missing_columns = [col for col in X_train.columns 
                   if X_train[col].isnull().any()]
```

```python
X_train_dropped = X_train.drop(missing_columns, axis = 1)
X_valid_dropped = X_valid.drop(missing_columns, axis = 1)
```

Score thr data set after removing columns


```python
score_dataset(X_train_dropped, X_valid_dropped, y_train, y_valid)
```

# 2nd approach is imputation

### Use the next code cell to impute missing values with the mean value along each column

```python
from sklearn.impute import SimpleImputer
```

```python
my_imputer = SimpleImputer()
```

```python
imputed_X_train= pd.DataFrame(my_imputer.fit_transform(X_train))
imputed_X_valid= pd.DataFrame(my_imputer.fit_transform(X_valid))
```

```python
imputed_X_train.columns=X_train.columns
imputed_X_valid.columns=X_valid.columns
```

```python
score_dataset(imputed_X_train, imputed_X_valid, y_train, y_valid)
```

# 3rd Approach


```python
final_imputer = SimpleImputer(strategy='median')
```

```python
final_X_train = pd.DataFrame(final_imputer.fit_transform(X_train))
final_X_valid  = pd.DataFrame(final_imputer.transform(X_valid))
```

```python
final_X_train.columns = X_train.columns
final_X_valid.columns = X_valid.columns
```

```python
model= RandomForestRegressor(n_estimators=100, random_state=0)
```

```python
model.fit(final_X_train,y_train)
```

```python
final_prediction= model.predict(final_X_valid)
```
